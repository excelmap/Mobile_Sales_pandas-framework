```python
import pandas as pd 
import numpy as np 
```

```python
data = pd.read_excel("Mobile Sales Data.xlsx")
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.info()
```

```python
# Data Cleaning
data.drop_duplicates()
```

```python
# Fixing the inconsistence of the data
data['Day Name'] = data['Day Name'].replace({'Sat':'Saturday', 'Sun':'Sunday', 'Mon':'Monday',
                          'Tue':'Tuesday', 'Wed':'Wednesday',
                          'Thu':'Thursday', 'Fri':'Friday'})
```

```python
data.head()
```

```python
# Combining Day, Month, Year in Date column
data['Date'] = pd.to_datetime(data[['Year', 'Month', 'Day']])
```

```python
# Now convert the date_added column to datetime
data['Date'] = pd.to_datetime(data['Date'], format='%B %d, %Y', errors='coerce')
data.info()
```

```python
data.drop(['Day', 'Month', 'Year'], axis=1, inplace=True)
data.dropna(inplace=True)
```

```python
# Define the rating_status function
def rating_status(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 3.5:
        return "Good"
    elif rating >= 2.5:
        return "Average"
    elif rating >= 1.5:
        return "Poor"
    else:
        return "Unknown"

# Applying function to the Customer Rating column
data['Customer Rating Status'] = data['Customer Ratings'].apply(rating_status)
```

```python
data.describe(include = 'number')
```

```python
# Calculate the Total Sales 
data['Total Revenue'] = data['Units Sold'] * data['Price Per Unit']
```

```python
# Calculate the total revenue per brand 
Total_Revenue_Per_Brand = data.groupby('Brand')['Total Revenue'].sum().sort_values(ascending=False)
Total_Revenue_Per_Brand
```

```python
# Sales by Payment Method 
Sales_payment_Method = data['Payment Method'].value_counts()

Sales_payment_Method
```

```python
# Calculate the average customer rating per brand 
Average_Rating_Brand = data.groupby('Brand')['Customer Ratings'].mean().sort_values(ascending=False)

Average_Rating_Brand
```

```python
# Calculate the average customer rating per Mobile Model
Average_Rating_Model = data.groupby('Mobile Model')['Customer Ratings'].mean().sort_values(ascending=False)

Average_Rating_Model
```

```python
# Calculate the most popular mobile model
Most_popular_Mobile_Model = data['Mobile Model'].value_counts().idxmax()

Most_popular_Mobile_Model
```

```python
# Calculate the unit sold by day name 
Total_Units_Day_Name = data.groupby('Day Name')['Units Sold'].sum().sort_values(ascending=False)

Total_Units_Day_Name
```

```python
# Calculate the Average Price Per Unit sold by brand 
Average_Price_Per_Unit_Brand = data.groupby('Brand')['Price Per Unit'].mean().sort_values(ascending=False)

Average_Price_Per_Unit_Brand
```

```python
# Calculate the Average Price Per Unit sold by Mobile Model
Average_Price_Per_Unit_Mobile_Model = data.groupby('Mobile Model')['Price Per Unit'].mean().sort_values(ascending=False)

Average_Price_Per_Unit_Mobile_Model
```
